```python
dotdash['cpm'] = np.where(dotdash['Total impressions'] > 0, 1000 * (dotdash['Revenue'] / dotdash['Total impressions']),
                          0)
```

```python
dotdash['viewability'] = np.where(dotdash['Total impressions'] > 0, 
                                  dotdash['Total Active View viewable impressions'] / dotdash['Total impressions'], 0)
```

```python
dotdash['cpm'].head(10)
```

```python
dotdash['viewability'].head(10)
```

```python
avg_cpm = dotdash['cpm'].mean()
print(avg_cpm)
```

```python
avg_view = dotdash['viewability'].mean()
print(avg_view)
```

```python
dotdash.groupby(dotdash['Date']).mean().sort_values(by = 'viewability', ascending = False)
```

```python
dotdash.groupby(dotdash['Date']).mean().sort_values(by = 'cpm', ascending = False)
```

#### Observations 
- After viewing the average CPM and Viewability for the dates (5/20/2018 - 6/2/2018), we see the the average **CPM has increased by .75** whereas the average **viewability has slightly dipped from its start at .55 to .53**
